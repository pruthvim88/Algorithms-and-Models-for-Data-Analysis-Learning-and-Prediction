# K-means clustering with Pew Data

```r
input_data <- read.csv("act3_data.csv",header = TRUE)
```

```r
head(input_data)
```

```r
kmeansdata <- data.frame(input_data$age,input_data$sex,input_data$marital,input_data$race,input_data$ideo,input_data$q1)
names(kmeansdata) <- c("Age","Sex","Marital.Status","Race","Political.Ideology","Community.Rate")
```

### Clustering data based on the how people rate their community along with various distributions based on Age, sex, Marital status, race and their political ideology

```r
head(kmeansdata)
```

```r
kmeansdata <- na.omit(kmeansdata)
```

```r
x = kmeansdata[,-7]
kc3 <- kmeans(x,3)
kc5 <- kmeans(x,5)
```

## 3 clusters

```r
kc3
```

```r
kc3$centers
```

# 5 Clusters

```r
kc5
```

```r
kc5$centers
```

## Sex distribution variation between 3 and 5 clusters

```r
Sex3 <- table(kmeansdata$Sex,kc3$cluster)
rownames(Sex3) <- c("Male","Female")
colnames(Sex3) <- c("Cluster1","Clsuter2","Clsuter3")
Sex3
```

```r
Sex5 <- table(kmeansdata$Sex,kc5$cluster)
rownames(Sex5) <- c("Male","Female")
colnames(Sex5) <- c("Cluster1","Clsuter2","Clsuter3","Clsuter4","Clsuter5")
Sex5
```

```r
plot(x[c("Sex","Community.Rate")], col=kc3$cluster)
points(kc3$centers[,c("Sex","Community.Rate")], col=1:3, pch=23, cex=3)
plot(x[c("Sex","Community.Rate")], col=kc5$cluster)
points(kc5$centers[,c("Sex","Community.Rate")], col=1:5, pch=23, cex=3)
```

## Marital people distribution variation between 3 and 5 clusters

```r
# marital Status cluster division
Marital3 <- table(kmeansdata$Marital.Status,kc3$cluster)
rownames(Marital3) <- c("Married","Living with a partner","Divorced", "Separated","Widowed","Never been married","Don't know","Refused")
colnames(Marital3) <- c("Cluster1","Clsuter2","Clsuter3")
Marital3
```

```r
# marital Status cluster division
Marital5 <- table(kmeansdata$Marital.Status,kc5$cluster)
rownames(Marital5) <- c("Married","Living with a partner","Divorced", "Separated","Widowed","Never been married","Don't know","Refused")
colnames(Marital5) <- c("Cluster1","Clsuter2","Clsuter3","Clsuter4","Clsuter5")
Marital5
```

```r
plot(x[c("Marital.Status","Community.Rate")], col=kc3$cluster)
points(kc3$centers[,c("Marital.Status","Community.Rate")], col=1:3, pch=23, cex=3)
plot(x[c("Marital.Status","Community.Rate")], col=kc5$cluster)
points(kc5$centers[,c("Marital.Status","Community.Rate")], col=1:5, pch=23, cex=3)
```
